### by Adam Richman

Good data visualization isn't just about looking attractive - it's about ensuring clarity, focus, and accessibility for everyone. 

Many users, especially those with ADHD, dyslexia, or cognitive processing disabilities, can struggle with overly complex or cluttered visuals. 

This notebook shows how small design improvements can dramatically improve accessibility.

## Modules

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
np.random.seed(42)
```

## Bad Charts

Here's a cluttered bar chart that suffers from many common cognitive accessibility issues:
- Too many colors
- Busy background
- Slanted labels
- Small fonts
- Hard-to-read legend
- Distracting gridlines
- Inconsistent line styles

This is the type of design that can cause cognitive overload.

```python
# data generation
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
products = ['Product A', 'Product B', 'Product C', 'Product D', 'Product E', 'Product F']
data = {product: np.random.randint(20, 100, size=12) for product in products}
df = pd.DataFrame(data, index=months)
```

```python
# plot
plt.figure(figsize=(12, 8))
styles = ['-', '--', '-.', ':', '-', '--']
colors = sns.color_palette("tab10", n_colors=6)
for idx, product in enumerate(products):
    plt.plot(df.index, df[product], label=product, linestyle=styles[idx], color=colors[idx], linewidth=2)

plt.grid(True)
plt.title('Monthly Sales (Messy Version)', fontsize=14)
plt.xlabel('Month', fontsize=10)
plt.ylabel('Sales', fontsize=10)
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

## Redisign for Cognitive Accessibility

Let's apply simple but powerful improvements:
- Use minimal, consistent colors
- Remove unnecessary gridlines and borders
- Straighten and enlarge labels
- Remove legend and directly label lines
- Increase white space
- Focus on the key trends only
- Simplify visual elements to minimize distractions

Simplicity improves focus.

```python
# focus on two products for simplicity
focus_products = ['Product A', 'Product B']
```

```python
# plot simpler version
plt.figure(figsize=(10, 6))

colors = ["#1f77b4", "#ff7f0e"]  # Simple high-contrast palette

for idx, product in enumerate(focus_products):
    plt.plot(df.index, df[product], label=product, color=colors[idx], linewidth=3)
    # Add direct text labels
    plt.text(11, df.iloc[-1][product], product, color=colors[idx], fontsize=12, va='center')

plt.title('Monthly Sales Trends (Accessible Version)', fontsize=18, weight='bold')
plt.xlabel('Month', fontsize=14)
plt.ylabel('Sales', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Reduce visual noise
plt.grid(False)
plt.box(False)

plt.tight_layout()
plt.show()
```

## Before and After Comparison

| Before | After |
|:---|:---|
| Many colors, confusing styles | Minimal colors, consistent lines |
| Slanted text, small fonts | Straight, readable labels |
| Reliance on legend | Direct labeling |
| Busy grid and borders | Clean, whitespace-focused design |

Good design isn't about adding more; it's about **removing obstacles**.

**Cognitive accessibility benefits everyone.**

## Additional Examples
Let's show more chart types redesigned for cognitive accessibility!

```python
# pie Chart vs simplified bar chart
labels = ['Category 1', 'Category 2', 'Category 3', 'Category 4', 'Category 5']
sizes = [20, 30, 15, 25, 10]

# pie chart (less accessible)
plt.figure(figsize=(7, 7))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, textprops={'fontsize': 10})
plt.title('Market Share (Pie Chart)', fontsize=16)
plt.show()

# bar chart (more accessible)
plt.figure(figsize=(8, 5))
colors = sns.color_palette("muted")
sns.barplot(x=sizes, y=labels, hue=labels, palette=colors[:5], legend=False)
plt.title('Market Share (Bar Chart)', fontsize=16, weight='bold')
plt.xlabel('Percentage')
plt.ylabel('Category')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

The bar chart allows:
- Easier comparison of values
- Clear labeling without needing to match colors
- Better scalability with more categories

```python
# redundant encoding (shape + color)
x = np.linspace(0, 10, 50)
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure(figsize=(10, 6))

plt.plot(x, y1, label='sin(x)', color='#1f77b4', marker='o', markersize=7)
plt.plot(x, y2, label='cos(x)', color='#ff7f0e', marker='s', markersize=7)

plt.title('Using Color + Shape for Redundancy', fontsize=18, weight='bold')
plt.xlabel('X-axis', fontsize=14)
plt.ylabel('Y-axis', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)

# direct labeling
plt.text(10.2, y1[-1], 'sin(x)', color='#1f77b4', fontsize=12, va='center')
plt.text(10.2, y2[-1], 'cos(x)', color='#ff7f0e', fontsize=12, va='center')

plt.box(False)
plt.tight_layout()
plt.show()
```

Adding shapes along with color ensures that people who cannot perceive color differences still distinguish the lines.

## Bonus Tips for Cognitive-Friendly Visuals

- Use high-contrast color combinations.
- Prefer sans-serif fonts (e.g., Arial, Helvetica) for better legibility.
- Limit the number of categories shown at once.
- Use consistent shapes, line styles, and symbols.
- Provide summaries alongside visuals to aid comprehension.
- Allow whitespace to "breathe" - don't fear empty space!
- Apply redundant encoding: pair color with shape, size, or labels.
